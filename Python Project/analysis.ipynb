# Visualizations

```python
df["Department"].value_counts().plot(kind="bar")

plt.title("Employees by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df["Age"], df["MonthlyIncome"])

plt.title("Age vs Monthly Income")
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.show()
```

```python
plt.hist(df["Age"], bins=10)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Attrition"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Attrition Distribution")
plt.ylabel("")
plt.show()
```

# Conclusion

- Most employees belong to a limited number of departments.
- Monthly income varies depending on the job role.
- Employee ages are mostly concentrated in the middle age groups.
- Attrition analysis helps identify the proportion of employees who left the company.
